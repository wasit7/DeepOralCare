```python
df_board = pd.read_csv('data/company-board.csv')
df_board
```

```python
df_share_holder = pd.read_csv('data/company-shareholder.csv')
df_share_holder
```

```python
_buck_company = []
for company in df_board.company.unique():
    _buck_company.append(Entity(**{
        'name': f"company/{company}",
        'attribute': {
            'name': company,
            'entity': 'company'
        }
    }))
Entity.objects.bulk_create(_buck_company)
```

```python
def get_or_create_person(row):
    entity, _ = Entity.objects.get_or_create(
        name=row.get('name', ""),
        defaults = {
            'attribute': {}
        }
    )
    data = entity.attribute
    if row.get('position', "") in data:
        entity.attribute[row.get('position', "")] = entity.attribute[row.get('position', "")] + [row.get('company')]
    else:
        entity.attribute[row.get('position', "")] = [row.get('company')]
    entity.save()
```

```python
_board_data = []
for index, item in df_board.iterrows():
    _board_data += [{
        'name': item.get('name', ""),
        'position': elm,
        'company': item.get('company', "")
    } for elm in item.get('positions', '').split(',')]
df_board2 = pd.DataFrame(_board_data)
df_board2
```

```python
df_board2.apply(lambda x: get_or_create_person(x), axis=1)
```

```python
def add_shared_holder(row):
    entity, _ = Entity.objects.get_or_create(
        name=row.get('name', ""),
        defaults = {
            'attribute': {}
        }
    )
    data = entity.attribute
    if "ผู้ถือหุ้น" in data:
        entity.attribute['ผู้ถือหุ้น'] = entity.attribute['ผู้ถือหุ้น'] + [row.get('company')]
    else:
        entity.attribute['ผู้ถือหุ้น'] = [row.get('company')]
    entity.save()
```

```python
df_share_holder.apply(lambda x: add_shared_holder(x), axis=1)
```
