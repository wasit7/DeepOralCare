```python
from py2neo import Graph
```

```python
graph = Graph("bolt://neo4j:7687", auth=("neo4j", "qwer1234"))
```

```python
data = graph.query("""
    match r=(s)-[p]->(o) 
    where s.name = "/c/en/person"
    return s,p,o limit 100
""")
```

```python
df = data.to_data_frame()
df['s']=df['s'].apply(lambda x: dict(x))
df['p']=df['p'].apply(lambda x: dict(x))
df['o']=df['o'].apply(lambda x: dict(x))
df
```

```python
df.to_dict('records')
```

```python
data = graph.query("""
    match r=(s)-[p]->(o) 
    where s.name = "/c/en/person"
    return s,p,o limit 100
""")
```

```python
data = data.data()
data
```

```python
def extract_value(elm):
    return {
        k: v.get('name',"")
        for k,v in elm.items()
    }
[extract_value(elm) for elm in data]
```
